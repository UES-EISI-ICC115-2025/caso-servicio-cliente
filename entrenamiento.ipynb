```python
# Importar librerías
from langchain_ollama import OllamaEmbeddings
from langchain.document_loaders import TextLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain.document_loaders import PyPDFLoader
```

```python
# Cargar documentos
# Puedes cargar PDFs o TXT
pdf_loader = DirectoryLoader("documentos/pdfs", glob="*.pdf", loader_cls=PyPDFLoader)  # carpeta con PDFs
# txt_loader = DirectoryLoader("documentos", glob="*.txt", loader_cls=TextLoader)

# docs = pdf_loader.load() + txt_loader.load()
docs = pdf_loader.load()

# Dividir en fragmentos
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)
```

```python
# Crear embeddings usando Ollama
# Se asume que Ollama está corriendo en otro servidor y accesible
embeddings = OllamaEmbeddings(model="llama3.2:1b", base_url="http://ollama-server:11434")  
```

```python
import chromadb

client = chromadb.HttpClient(host="chroma-server", port=8000, ssl=False)

# Conectar al Chroma remoto
vector_store = Chroma(
    client=client,
    embedding_function=embeddings,
    collection_name="manuales_y_procedimientos",
)

# Indexar los fragmentos
vector_store.add_documents(chunks)

# Listo, ahora tus vectores están en Chroma remoto
print("Vector store creado y poblado en Chroma DB remoto.")
```

```python
query = "¿Cómo reiniciar el sistema?"
results = vector_store.similarity_search(query, k=1)

for r in results:
    print(r.page_content)
```
