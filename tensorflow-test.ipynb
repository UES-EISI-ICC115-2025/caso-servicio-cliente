```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

- **TensorFlow (tf)**: el framework principal de Machine Learning.
- **Keras**: API de alto nivel dentro de TensorFlow para construir y entrenar modelos.
- **NumPy (np)**: librería para manejar arreglos y operaciones numéricas.

1. Data

Creamos un dataset muy simple:
- Entradas x: `0, 1, 2, 3`
- Salidas y: `0, 1, 2, 3`

La función que queremos que el modelo aprenda es: **y = x**

```python
x = np.array([[0], [1], [2], [3]], dtype=float)
y = np.array([[0], [1], [2], [3]], dtype=float)
```

2. Construcción del modelo

- `Sequential`: indica que las capas se apilan en orden.

- `Dense(units=1)`: una neurona, produce una salida.

- `input_shape=[1]`: la entrada es un número por ejemplo.

En otras palabras, este modelo es equivalente a una función lineal: **y = w·x + b**

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

3. Compilación del model
- `optimizer='sgd'`: usa Stochastic Gradient Descent para ajustar los pesos.

- `loss='mean_squared_error'`: mide el error promedio entre la predicción y el valor real.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

3. Entrenamiento
- `fit`: entrena el modelo con los datos.
- `epochs=50`: repite el entrenamiento 50 veces sobre el dataset.

Durante el entrenamiento, TensorFlow ajusta los pesos w y el sesgo b para que las predicciones se acerquen a los valores reales.

```python
model.fit(x, y, epochs=50)
```

4. Predicción

Aquí pedimos al modelo: *Si x = 10, ¿qué valor de y predices?*

El modelo debería responder con un valor cercano a 10, ya que aprendió la relación **y ≈ x**.

```python
print(model.predict(np.array([10.0])))
```
